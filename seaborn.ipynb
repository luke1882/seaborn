Seaborn является надстройкой над Matplotlib.
Основная идея Seaborn заключается в предоставлении высокоуровневых команд для создания различных графических типов, используемых при исследовании статистических данных.

Рассмотрим пример изображения данных при помощи Matplotlib:

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('classic')

#создаём случайные данные
rng = np.random.RandomState(0)
x = np.linspace(0, 10, 500)
y = np.cumsum(rng.randn(500, 4), 0)

plt.subplots(figsize=(12,6))

#создаём изображение
plt.plot(x, y)

#настройка легенды
plt.legend('ABCD', ncol=2, loc='upper left')

plt.show();
```

Теперь сделаем то же самое, используя Seaborn:

```python
import seaborn as sns

#устанавливаем параметры изображения Seaborn по умолчанию
sns.set()

#задаём размер изображения
plt.subplots(figsize=(12,6))

#создаём изображение
plt.plot(x, y)

#настройка легенды
plt.legend('ABCD', ncol=2, loc='upper left')

plt.show();
```

Рассмотрим поближе некоторые функции Seaborn.

Функция sns.set() позволяет настроить параметры изображения:

```python
#пример использования функции sns.set()

import matplotlib.pyplot as plt
import seaborn as sns

#параметры функции sns.set():
#context : Масштаб изображения: 'paper', 'notebook', 'talk', 'poster'
#style : Фон: 'darkgrid', 'whitegrid', 'dark', 'white', 'ticks'
#palette : Цветовая гамма : 'deep', 'muted', 'bright', 'pastel', 'dark', 'colorblind'
#font : Шрифт : 'serif', 'sans-serif', 'fantasy', 'monospace', 'sans'
#font_scale : Размер шрифта (в виде множителя)
sns.set(context='notebook', style='whitegrid', palette='muted', font='sans-serif', font_scale=1.06)

#задаём размер изображения
plt.subplots(figsize=(12,6))

#создаём изображение
plt.plot(x, y)

#настройка легенды
plt.legend('ABCD', ncol=2, loc='upper left')

plt.show();
```

Функция sns.kdeplot() отвечает за изображение распределения плотностей значений базы данных в виде гладкой линии:

```python
#пример использования функции sns.kdeplot()

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

#создаём данные в формате двумерного массива
data = np.random.rand(5000,2)
data = pd.DataFrame(data, columns=['A', 'B'])

#настройка параметров изображения
sns.set(context='notebook', style='whitegrid', palette='muted', font='sans-serif', font_scale=1.06)

#задаём размер изображения
plt.subplots(figsize=(12,6))

#параметры функции sns.kdeplot()
#shade : Затенить область под графиком : True, False
#vertical : Расположить плотности на оси абсцисс : True, False
#kernel : Различные формы распределения : 'gau', 'cos', 'biw', 'epa', 'tri', 'triw'
#gridsize : Число точек для выборки : число
#legend  : Добавить легенду : True, False
sns.kdeplot(data['A'], shade=True)
sns.kdeplot(data['B'], shade=False, kernel='cos', gridsize=200, legend=True)

#Настройка легенды
plt.legend('AB', ncol=2, loc='upper left');
```

Функция sns.kdeplot() может принимать на вход 2 одномерных массива.
В этом случае отображается распределение точек 2-мерного массива на плоскости:

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

#настройка параметров изображения
sns.set(context='notebook', style='whitegrid', palette='bright', font='sans-serif', font_scale=1.06)

#задаём размер изображения
plt.subplots(figsize=(12,6))

#cbar : Показать шкалу соответствия плотностей и цветов : True, False
sns.kdeplot(data['A'], data['B'], cbar=True);
```

Функция sns.distplot() позволяет увидеть плотности в виде гистограммы и/или линии:

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

plt.subplots(figsize=(12,6))

#параметры функции sns.distplot()
#hist : Добавление гистограммы на график : True, False
#kde : Добавление KDE на график : True, False
#color : Цвет графика : 'Blue', 'Red', 'Green' и т.д.
#vertical : Расположить плотности на оси абсцисс : True, False
#label : Название кривой : 'название'
#axlabel : Название оси значений : 'название', 'False'
sns.distplot(data['A'], color='Red', label='A')
sns.distplot(data['B'], color='Green', label='B', axlabel='ОСЬ ЗНАЧЕНИЙ')

#настройка легенды
plt.legend('AB', ncol=2, loc='upper left');
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

#подгружаем 5-мерную базу данных Iris.csv
iris = pd.read_csv('D://Загрузки/iris.csv', engine='python')
print('')
print('Структура нашей базы: \n\n', iris.head())

#параметры функции sns.pairplot()
#hue : Переменная (столбец) базы данных, по которому будет сделана раскраска: 'название'
#hue_order: Порядок названий для раскраски (из столбца для раскраски) : список названий
#palette: Цвета для раскраски : словарь {'имя' : 'цвет'}
#vars : Список переменных (столбцов) базы для использования на графиках : список имён (по умолчанию - все)
#kind : Формат графика для отношений типа x -- y : 'scatter', 'reg'
#diag_kind : Формат графика для отношений типа x -- x : 'auto', 'hist', 'kde'
#height : Высота каждого графика
#aspect : Ширина каждого графика = aspect * height
sns.pairplot(iris, hue='вид', hue_order=('versicolor','setosa','virginica'), 
             palette={'versicolor':'Red','setosa':'Green','virginica':'Blue'}, diag_kind='hist', height=3);
```

Есть множество функций, дающих разные типы графиков: 
sns.pointplot(), sns.barplot(), sns.stripplot(), sns.swarmplot(), sns.boxplot(), sns.violinplot(), sns.boxenplot()
Работают как с базами формата pd.DataFrame(), так и с pd.Series(), np.array() и обычными списками.

Функция sns.catplot() объединяет их все. 
Отличие состоит в том, что sns.catplot() работает только с базами формата pd.DataFrame().

Чтобы нарисовать график определённого типа, нужно просто задать значение параметра kind:

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

#подгружаем 5-мерную базу данных Iris.csv
iris = pd.read_csv('D://Загрузки/iris.csv', engine='python')

sns.set(font_scale=1.5)

#параметры функции sns.catplot()
#x : Столбец базы для оси OX (числовой) : 'название_столбца'
#y : Столбец базы для оси OY (числовой) : 'название_столбца'
#hue : Столбец базы для раскраски : 'название_столбца'
#data : База данных (в формате pd.DataFrame()) : база
#row : Разбить график на подграфики по значениям столбца базы (по строкам) (аналог sns.pairplot()) : 'название_столбца'
#col : Разбить график на подграфики по значениям столбца базы (по столбцам) (аналог sns.pairplot()) : 'название_столбца'
#col_wrap : Нарисовать графики для каждого значения столбца col, в одной строке по col_wrap графиков : число
#order : Порядок столбцов базы: список названий
#hue_order : Порядок элементов столбца для раскраски : список названий
#col_order : Список столбцов для col_wrap : список названий
#kind : Тип графика : 'point', 'bar', 'strip', 'swarm', 'box', 'violin', 'boxen'
#height : Высота графика
#aspect : Ширина графика = aspect * height
#orient : Ориентация графика : 'v', 'h'
#color : Цвет графика : 'Blue', 'Red', 'Green' и т.д.
#palette : Цветовая гамма : 'deep', 'muted', 'bright', 'pastel', 'dark', 'colorblind'
#legend  : Добавить легенду : True, False

#функция sns.catplot() с параметрами x, y, hue
sns.catplot(x='длина_чашечки', y='ширина_чашечки', hue='вид', data=iris, 
            hue_order=['virginica','versicolor','setosa'], kind='strip', height=15);

#функция sns.catplot() с параметрами col, col_wrap (сравнение видов setosa и virginica)
sns.catplot(col='вид', data=iris, kind='strip', col_wrap=2, col_order=['setosa', 'virginica'], 
            order=['длина_чашечки', 'ширина_чашечки', 'длина_лепестка', 'ширина_лепестка'], height=10);
```

Имеется база данных доходов жителей Сан-Франциско за 2011-2014 гг. в файле salaries.csv
Найдём распределение плотностей этих зарплат:

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

#подгружаем базу зарплат SF из файла salaries.csv
salaries = pd.read_csv('D://Загрузки/salaries.csv', engine='python')

print('Структура нашей базы: ')
print(salaries.head())

#количество строк в базе
rng = len(salaries['Id'])
rows_for_drop = []

#найдём строки, в которых зарплата содержит символы, отличающиеся от цифр и точки
#выражение salaries.at[row, column] позволяет изменить значение в базе pd.DataFrame()
#выражение salaries.drop(list_of_rows) удаляет из базы salaries строки с номерами из списка list_of_rows
for k in range(rng):    
    if not str(salaries.at[k, 'TotalPay']).replace('.','').isnumeric():
            rows_for_drop.append(k)

print('')
print('Число строк в базе до чистки: ', len(salaries))
            
#удаляем из базы строки с некорректными зарплатами
salaries = salaries.drop(rows_for_drop)

print('Число строк в базе после чистки: ', len(salaries))

#задаём размер рисунка
plt.subplots(figsize=(15,7))

#создаём график распределения плотностей
sns.distplot(salaries['TotalPay'], color='Green', label='Зарплата', axlabel='Зарплаты жителей SF')

#настройка легенды
plt.legend(['Зарплата'], ncol=2, loc='upper right');
```
