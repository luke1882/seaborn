Сотни тысяч людей в разное время заявляли, будто видели корабль пришельцев, а некоторые вступили с ними в контакт.

В файле ufo.csv зафиксировано более 80 тысяч таких случаев за период 1949-2014 гг.

Отметьте на 'карте' контакты жителей Индии с неземной цивилизацией (в виде точек на 'карте')
('карта' - это координатная плоскость, где OX - долгота, а OY - широта).
Узнайте, какой формы были НЛО, посетившие эту страну. (точки разных цветов, цвет - это форма корабля)
Постройте распределение плотностей появления инопланетян по местности и выясните, какие города Индии представляют для пришельцев наибольший интерес. 

географические координаты Индии: https://yandex.ru/maps/
